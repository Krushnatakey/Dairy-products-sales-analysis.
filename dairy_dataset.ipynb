```python
df['Date']=pd.to_datetime(df['Date'])
df['Expiration_Date']=pd.to_datetime(df['Expiration_Date'])
```

```python
df=df.astype({'Quantity_in_Stock_liters_kg':'float','Quantity_Sold_liters_kg_':'float','Shelf_Life_days':'float'})
```

```python
df.dtypes
# changed data type
```

```python
y=[]
for i in range(len(df)):
    x=tuple(df.iloc[i])
    y.append(x)
    
y    
```

```python
file=open('dairy products','w')
for tuple in y:
    file.write(str(tuple)+','+'\n')
file.close()
```
